```python
import face_recognition

picture_of_me = face_recognition.load_image_file("HP1.jpg")
my_face_encoding = face_recognition.face_encodings(picture_of_me)[0]

# my_face_encoding now contains a universal 'encoding' of my facial features that can be compared to any other picture of a face!

unknown_picture = face_recognition.load_image_file("HP2.jpg")
unknown_face_encoding = face_recognition.face_encodings(unknown_picture)[0]

# Now we can see the two face encodings are of the same person with `compare_faces`!

results = face_recognition.compare_faces([my_face_encoding], unknown_face_encoding)

if results[0] == True:
    print("It's a picture of Harry Potter!")
else:
    print("It's not a picture of Harry Potter!")
```

```python
import face_recognition
import cv2
import matplotlib.pyplot as plt

# Load the image
image_path = "HP1.jpg"
image = cv2.imread(image_path)

# Convert the image from BGR to RGB (as face_recognition works with RGB images)
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Detect face landmarks
face_landmarks_list = face_recognition.face_landmarks(rgb_image)

# Draw landmarks on the image
for face_landmarks in face_landmarks_list:
    for facial_feature in face_landmarks.keys():
        for point in face_landmarks[facial_feature]:
            cv2.circle(image, point, 2, (0, 255, 0), -1)  # Draw a green circle at each landmark point

# Display the image with landmarks
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Turn off axis
plt.show()
```
